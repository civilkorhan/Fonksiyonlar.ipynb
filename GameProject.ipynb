```python
import random
class asker:
    def __init__(self,isim,can,SaldiriGucu,mermi):
        self.isim=isim
        self.can=can
        self.SaldiriGucu=SaldiriGucu
        self.mermi=mermi


    def saldiri(self):
        print(self.isim+" :SALDIRIYORUUUMMM")
        harcanan_mermi=random.randrange(10,20)
        print(str(harcanan_mermi)+" tane mermi harcadım")
        #self.can=self.can-30
        self.mermi-=harcanan_mermi
        return(harcanan_mermi,self.SaldiriGucu)

    def savunma(self,harcanan_mermi,SaldiriGucu):
        print(self.isim+" :SALDIRIYA UĞRUYORUMMM")
        self.can-= (harcanan_mermi*SaldiriGucu)
        print("Vuruldummm")
    def basiliyor(self):
        print("Basılıyorr")
        print("İsim: "+self.isim,"Can: "+str(self.can),"Güç: "+str(self.SaldiriGucu),"Mermi: "+str(self.mermi))
    def mermi_yeterli_mi(self):
        if (self.mermi>=30):
            print(self.isim + " :Konuşuyor.Mermim Yeterli")

            return True
        return False

askerler=[]

i=0
while (i<10):
    random_can=random.randrange(50,100)
    random_SaldiriGucu=random.randrange(5,10)
    random_mermi=random.randrange(30,100)
    yeniAsker=asker("Asker"+str(i+1),random_can,random_SaldiriGucu,random_mermi)
    askerler.append(yeniAsker)
    i+=1
for asker in askerler:
    asker.basiliyor()



i=0
while(i<3):
    random_saldiran_asker = random.randrange(1,10)
    random_savunan_asker = random.randrange(1,10)

    donen_deger=askerler[random_saldiran_asker].saldiri() # bu dışarıya return olarak(20,5)gibi bir tuple çıkacak.
    askerler[random_savunan_asker].savunma(donen_deger[0],donen_deger[1])
    askerler[random_saldiran_asker].mermi_yeterli_mi()

    i+=1
    print("Round Bitti "+str(i))
```
