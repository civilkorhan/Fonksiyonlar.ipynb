```python
liste1=[1,2,3,4,5,6]
def square_Sum1(liste):
  result=0
  for i in liste:
    result=result+i**2
  return result
square_Sum1(liste1)


```

```python
def square_Sum2(n):
 result=0
 for i in range(1,n+1):
   result=result+i**2
 return result

square_Sum2(6)
```

```python
def square_Sum3(n):
  return n*(n+1)*(2*n+1)/6
square_Sum3(6)
```

```python
%timeit square_Sum3(5)
%timeit square_Sum2(5)



```

Big O Notation

```python
#Big O Notation Kod Bloklarının maliyetini performansını hesaplar.
#Big O bir fonksiyonun büyümesini ifade eder.
#Amaç operasyonu azaltmak kod verimini artırmak,kod tekrar tekrar operasyon yapsın istenmez.
#Big O en kötü kullanım, Omega en iyi kullanım Tetha ise ortalama diyebileğimiz kullanımdır.
#Input Size dahi Time Complexy'i etkiler.
#Constant sabit operasyon Time Complexlty'i etkilemez yapılacak iş aynıdır.Örnek vermek gerekirse 3 indexli ve 10000 indexli bir dizi de ilk değeri yazdırma metotu bellekte aynı yer kaplar.
#Linear O da input sayısı Time Complexlty'i etkiler.
#Kaynak:Ömer Cengiz sevgilerle..
```

```python
#Constant 0(1)
def constant_BigO(list):
  print(list[0])
constant_BigO([1,2,3,4,5])
```

```python
#Linear o(n)
def linear_BigO(list):
  for i in list:
    print(i)
linear_BigO([1,2,3,4,5])
```
